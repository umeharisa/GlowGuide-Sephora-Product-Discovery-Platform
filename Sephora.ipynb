**Pre-processing of the review text**

```python
import pandas as pd
import numpy as np  # Import numpy for NaN handling
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string

# Download NLTK resources (if not already downloaded)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Text preprocessing function
def preprocess_text(text):
    if isinstance(text, str):  # Check if text is not NaN
        # Step 1: Lowercasing
        text = text.lower()

        # Step 2: Tokenization
        words = word_tokenize(text)

        # Step 3: Removing Special Characters and Numbers
        words = [word for word in words if word.isalpha()]

        # Step 4: Stopword Removal
        stop_words = set(stopwords.words('english'))
        words = [word for word in words if word not in stop_words]

        # Step 5: Lemmatization (or stemming if preferred)
        lemmatizer = WordNetLemmatizer()
        words = [lemmatizer.lemmatize(word) for word in words]

        # Join the processed words back into a clean text
        cleaned_text = ' '.join(words)
    else:
        cleaned_text = ''  # Handle missing values by setting an empty string

    return cleaned_text

# Apply the preprocessing function to the 'review_text' column
df_selectedCols['cleaned_review'] = df_selectedCols['review_text'].apply(preprocess_text)

# Print the DataFrame with cleaned text
print(df_selectedCols[['review_text', 'cleaned_review']])
```

**Sentiment analysis of cleaned text**

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Initialize the VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Function to get sentiment polarity
def get_sentiment_polarity(text):
    sentiment_scores = analyzer.polarity_scores(text)
    compound_score = sentiment_scores['compound']
    if compound_score >= 0.05:
        return 'Positive'
    elif compound_score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Apply sentiment analysis to the 'review_text' column
df_selectedCols['sentiment'] = df_selectedCols['cleaned_review'].apply(get_sentiment_polarity)

# Print the DataFrame with sentiment analysis results
print(df_selectedCols[['cleaned_review', 'sentiment']])
```

```python
# Calculate the mean rating per product and store it in 'avg_rating' column
df_selectedCols['avg_rating'] = df_selectedCols.groupby('product_name')['rating'].transform('mean')

# Round the 'avg_rating' column to two decimal places
df_selectedCols['avg_rating'] = df_selectedCols['avg_rating'].round(2)

# Check for null values in the DataFrame after rounding
print(df_selectedCols.isnull().sum())
```

## Identifying Skin Care Concerns from Customer Reviews

In this code, we are analyzing customer reviews related to skincare products and identifying specific skincare concerns based on keywords mentioned in the reviews.

1. **Defining a Dictionary of Concerns and Keywords:**
   - We start by creating a dictionary called `concern_keywords`. This dictionary maps various skincare concerns to associated keywords. For example, the concern "Acne" is associated with keywords like "acne," "pimple," "breakout," and "blemish."

2. **Creating a Function to Determine Concerns:**
   - We define a function called `get_concern` that takes a "cleaned_review" as input.
   - Inside this function, we initialize an empty list called "concerns" to store the identified skincare concerns.
   - The function checks each review for keywords related to different concerns by iterating through the `concern_keywords` dictionary.
   - If it finds any keywords in the review, it adds the corresponding concern to the "concerns" list.
   - Finally, the function returns a comma-separated list of concerns found in the review or "Other" if no concerns are identified.

3. **Applying the Function to Create a 'concern' Column:**
   - We apply the `get_concern` function to the "cleaned_review" column of the DataFrame, which contains the cleaned customer reviews.
   - The results are stored in a new column called "concern."

4. **Viewing the DataFrame:**
   - We print the DataFrame, displaying both the "cleaned_review" and "concern" columns.

This code allows us to automatically categorize skincare concerns based on customer reviews, which can be valuable for product analysis and customer feedback analysis.

```python
# Function to determine the concern(s) based on keywords in the cleaned review
def get_concern(cleaned_review):
    concerns = []  # Initialize an empty list to store concerns
    for concern, keywords in concern_keywords.items():
        for keyword in keywords:
            if keyword in cleaned_review:
                concerns.append(concern)
    return ', '.join(concerns) if concerns else 'Other'  # Comma-separated concerns or 'Other' if none found

# Apply the get_concern function to create a 'concern' column
df_selectedCols['concern'] = df_selectedCols['cleaned_review'].apply(get_concern)

# Print the DataFrame with the 'concern' column
print(df_selectedCols[['cleaned_review', 'concern']])
```

```python
# Define a custom function to remove duplicate words within a concern 

def remove_duplicate_words(concern):
    # Split the string into words
    words = concern.split(', ')

    # Remove duplicates by converting the list of words to a set and back to a list
    unique_words = list(set(words))

    # Join the unique words back into a string
    cleaned_concern = ', '.join(unique_words)

    return cleaned_concern

# Apply the custom function to each row in the 'concern' column
df_selectedCols['concern'] = df_selectedCols['concern'].apply(remove_duplicate_words)
```

The code filters a DataFrame of skincare concerns, excluding entries labeled as 'other.' It then counts and ranks the unique concerns based on their frequency, identifying the top 100 most common skincare concerns. This summary can assist businesses and researchers in understanding prevalent customer concerns and trends within skincare product reviews.

```python
import pandas as pd

# Assuming you have a DataFrame named df with a 'concern' column

# Step 1: Filter out rows with 'other' concern
filtered_df = df_selectedCols[df_selectedCols['concern'] != 'other']

# Step 2: Get the counts of each unique concern in the filtered DataFrame
concern_counts = filtered_df['concern'].value_counts()

# Step 3: Sort the concerns based on their counts to find the top concerns
top_100_concerns = concern_counts.head(100).index.tolist()  # Convert to a list of strings

# Create a new DataFrame to store the top 100 concerns
top_100_concerns_df = pd.DataFrame({'Top Concerns': top_100_concerns})

# Print the new DataFrame
print(top_100_concerns_df)
```

The code begins by creating a TF-IDF (Term Frequency-Inverse Document Frequency) vectorizer to transform the top 100 skincare concerns into numerical vectors, emphasizing the importance of unique terms across concerns. Agglomerative Clustering is chosen as the clustering algorithm to group these concerns into 13 clusters.

**Agglomerative Clustering** is suitable for this problem because it hierarchically merges concerns, creating a tree-like structure of clusters. This method is useful when there's no predefined number of clusters, allowing for flexible clustering based on the similarity of concerns. The code subsequently extracts unique concerns within each cluster, removing duplicates and providing a summary of distinct groups of related skincare concerns. This approach aids in identifying common themes and trends among a large number of concerns.

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.feature_extraction.text import TfidfVectorizer

# Assuming you have a DataFrame named top_47_concerns_df with a 'Top Concerns' column

# Create a TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Vectorize the top concerns
tfidf_matrix = tfidf_vectorizer.fit_transform(top_100_concerns_df['Top Concerns']).toarray()  # Convert to a dense array

# Define the number of clusters
n_clusters = 13

# Create an instance of the AgglomerativeClustering model
model = AgglomerativeClustering(n_clusters=n_clusters)

# Fit the model to your dense TF-IDF data
cluster_labels = model.fit_predict(tfidf_matrix)

# Create an empty dictionary to store clusters
clusters = {}

# Iterate through each cluster label and store concerns
for cluster_id in range(n_clusters):
    # Extract concerns for the current cluster
    cluster_contents = top_100_concerns_df['Top Concerns'][cluster_labels == cluster_id].tolist()

    # Remove duplicates from the list of concerns within each cluster
    unique_concerns = list(set(cluster_contents))

    # Store the unique concerns in the dictionary
    clusters[cluster_id] = unique_concerns

# Print the concerns in each cluster
for cluster_id, unique_concerns in clusters.items():
    print(f"Cluster {cluster_id}: {unique_concerns}")
```

**finding the best products based on positive sentiment and highest ratings**

```python
import pandas as pd
import pickle
from sklearn.metrics.pairwise import linear_kernel
from sklearn.feature_extraction.text import TfidfVectorizer

# Create a TF-IDF vectorizer to convert text to numerical features
tfidf_vectorizer = TfidfVectorizer(stop_words='english')

# Filter products with positive sentiment
positive_products = df_selectedCols[df_selectedCols['sentiment'] == 'Positive']

# Fit and transform the 'concerns' column to obtain TF-IDF features for positive sentiment products
tfidf_matrix_positive = tfidf_vectorizer.fit_transform(positive_products['concern'])
```

Transform user's concerns to TF-IDF features

```python
# Example user concerns
user_concerns = 'dry lips'

# Transform the user concerns to TF-IDF features
user_concerns_vector = tfidf_vectorizer.transform([user_concerns])

# Calculate cosine similarity between user's concerns and product concerns for positive sentiment products
cosine_similarities = linear_kernel(user_concerns_vector, tfidf_matrix_positive)

# Get the product indices sorted by similarity
product_indices = cosine_similarities.argsort()[0][::-1]
```

Recommend top products based on highest ratings

```python
# Assuming 'positive_products' DataFrame contains the positive sentiment products sorted by average rating

# Initialize variables to keep track of recommended products and their details
recommended_products = []
recommended_product_names = set()  # Using a set to track product names for uniqueness
count = 0
top_n = 5  # Number of recommendations to provide

# Loop through product indices and recommend top N products with the highest ratings
for idx in product_indices:
    product_name = positive_products['product_name'].iloc[idx]
    price = positive_products['price_usd'].iloc[idx]
    average_rating = positive_products['avg_rating'].iloc[idx]

    # Check if the product name is not already recommended to avoid duplicates
    if product_name not in recommended_product_names:
        recommended_products.append({
            'product_name': product_name,
            'price': price,
            'average_rating': average_rating
        })

        recommended_product_names.add(product_name)  # Add product name to the set
        count += 1

    if count == top_n:
        break


# Print recommended products with details
print("Recommended Products based on positive sentiment and highest ratings:")
for product in recommended_products:
    print(f"Product Name: {product['product_name']}, Price: ${product['price']}, Average Rating: {product['average_rating']}")
```
